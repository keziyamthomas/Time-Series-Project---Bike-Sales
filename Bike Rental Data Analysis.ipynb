```python
# importing necessary packages
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bike_rental_df = pd.read_csv("Dataset/BikeRentalData.csv",header=0,delimiter=";")
```

```python
bike_rental_df.head()
```

```python
bike_rental_df = bike_rental_df[["Datetime","Measure","#Value"]]
```

```python
bike_rental_df.head()
```

```python
bike_rental_pivot_df = bike_rental_df.pivot(index='Datetime', columns='Measure', values='#Value')
```

```python
bike_rental_pivot_df.head()
```

```python
bike_rental_selected_df = bike_rental_pivot_df.rename_axis(None, axis=1)
```

```python
bike_rental_selected_df.head()
```

```python
bike_rental_selected_df.plot.line(figsize=(20,6))
```

```python
bike_rental_final_df = bike_rental_selected_df.reset_index()
```
